```python
def voigt(x, x0, A, sigma, gamma, m, b):
    return A*wofz((x-x0+1j*gamma)/sigma/np.sqrt(2)).real + m*x + b

def gaussian(x, x0, A, sigma, m, b):
    return A*np.exp(-(x-x0)**2/2/sigma**2) + m*x+b

def gaussian2(x,x1,x2,A1,A2,sigma,m,b):
    return A1*np.exp(-(x-x1)**2/2/sigma**2) + A2*np.exp(-(x-x2)**2/2/sigma**2) + m*x + b
```

```python
width = 20

def segment(element, lam0, width=width):

    x, y = element['wavelength'], element['signal'] 
    idx = np.where((x>lam0-width/2) & (x<lam0+width/2))    
    lam_max = x[idx][np.argmax(y[idx])]    
    idx = np.where((x>lam_max-width/2) & (x<lam_max+width/2))
    centroid = np.sum(y[idx]*x[idx])/np.sum(y[idx])
    x_c, y_c = x[np.argmin(np.abs(x-centroid))], y[np.argmin(np.abs(x-centroid))]
    idx = np.where((x>x_c-width/2) & (x<x_c+width/2))
    
    return x[idx],y[idx]

def segment2(element, lam0, width=2.5*width):
    
    x, y = element['wavelength'], element['signal']
    idx = np.where((x>lam0-width/2) & (x<lam0+width/2))
    x = x[idx]
    y = y[idx]
    
    return x,y


def fit(element, lam0, A):
    
    x, y = segment(element, lam0)
    params = curve_fit(gaussian, x, y, p0=[lam0, A, 10,0,0])
    
    return params[0], np.sqrt(np.diag(params[1]))


def fit2(element, lam1, lam2, A1, A2):
    
    x, y = segment2(element, (lam1+lam2)/2)
    params = curve_fit(gaussian2, x, y, p0=[lam1, lam2, A1, A2, 10,0,0])
    
    return params[0], np.sqrt(np.diag(params[1]))
```

```python
t = 8.11*u.mm
def energy(r, r1, r2):
    return (con.h*con.c/2/t * np.abs((r**2 - r1**2)/(r1**2 - r2**2))).to(u.erg)
```

```python
nsample = 10000

all_energy = np.zeros((nsample, 4))
all_field = np.zeros((nsample, 4))
mub = np.zeros(nsample)
for i in range(nsample):
    r1 = 39 + np.random.normal(0, 0.5)
    r2 = 45 + np.random.normal(0, 0.5)
    pos_arr = np.array([0, 41, 42.5, 43.5]) + np.random.normal(0, 0.5)
    pos_arr[0] = r1
    energy_arr = energy(pos_arr, r1, r2)
    all_energy[i] = energy_arr

    current = np.array([0,0.7,1.4,2]) + np.random.normal(0, 0.05)
    
    slope = bfit[0] + np.random.normal(0, np.sqrt(np.diag(bcov))[0])
    offset = bfit[1] + np.random.normal(0, np.sqrt(np.diag(bcov))[1])
    
    field = (slope*current + offset)*1e3
    all_field[i] = field
    
    fit = curve_fit(lambda x, a: a*x, xdata=field, ydata=energy_arr)
    
    mub[i] = fit[0][0]
```
