```python
import numpy as np
```

```python
def Num(a,b,N):
    return(N/2*np.cos(b-a))
```

```python
anglesa = np.array([0,0,45,90])/180*np.pi
anglesb = np.array([0,90,45,90])/180*np.pi
```

```python
Num(anglesa,anglesb,1000)
```

```python
def Num1(a,b,N):
    return(N*(np.sin(a)*np.sin(b)*np.sqrt(3)/2+np.cos(a)*np.cos(b)))
```

```python
Num1(anglesa,anglesb,1000)
```

```python
def Num2(a,b,N):
    return(N/4*(np.sin(a)*np.sin(b)-np.sin(a)*np.cos(b)+np.cos(a)*np.cos(b)-np.cos(a)*np.sin(b))**2)
```

```python
Num2(anglesa,anglesb,1000)
```
